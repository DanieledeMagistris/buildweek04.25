```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("prodotti mizuno unito.csv")
df
```

```python
df[["modello","sport","colori" ]]=df["nome_prodotto"].str.split("-", expand=True)
df
```

```python
df["prezzo_effettivo"] = df["prezzo_effettivo"].str.replace(",", ".").astype(float)
df["prezzo_ordinario"] = df["prezzo_ordinario"].str.replace(",", ".").astype(float)
```

```python
df
```

```python
raggruppamento=df.groupby("sport")[["prezzo_effettivo", "prezzo_ordinario"]].sum()
```

```python
raggruppamento = raggruppamento.sort_values(by=["prezzo_effettivo", "prezzo_ordinario"])
raggruppamento
```

```python
plt.figure(figsize=(10, 6))

# Barre per il prezzo effettivo
plt.barh(df["sport"], df["prezzo_effettivo"], color="skyblue", label="Prezzo Effettivo")

# Barre per il prezzo ordinario
plt.barh(df["sport"], df["prezzo_ordinario"], color="orange", label="Prezzo Ordinario", alpha=0.7)

# Aggiungi etichette e titolo
plt.xlabel("Prezzo (€)")
plt.ylabel("Sport")
plt.title("Confronto Prezzo Effettivo e Prezzo Ordinario per Sport")
plt.legend()

# Mostra il grafico
plt.tight_layout()
plt.show()
```
